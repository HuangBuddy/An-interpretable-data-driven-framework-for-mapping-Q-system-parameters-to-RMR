```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl
from matplotlib.ticker import MultipleLocator

mpl.rcParams["font.family"] = "Times New Roman"
mpl.rcParams["font.serif"] = ["Times New Roman"]
mpl.rcParams["pdf.fonttype"] = 42
mpl.rcParams["ps.fonttype"] = 42
mpl.rcParams["axes.unicode_minus"] = False

file_path = "Pop_Compare.xlsx"
df = pd.read_excel(file_path)

x_col = "Iteration"
y_cols = ["Population-20", "Population-50", "Population-100", "Population-150", "Population-200"]

fig, ax = plt.subplots(figsize=(8, 5))

for col in y_cols:
    ax.plot(df[x_col], df[col], linewidth=2, label=col)

ax.set_xlabel("Iteration", fontsize=14, fontname="Times New Roman")
ax.set_ylabel("Fitness (Validation RMSE)", fontsize=14, fontname="Times New Roman")
ax.set_title("Convergence curves under different population sizes", fontsize=14, fontname="Times New Roman")

ax.set_ylim(1.55, 2.05)
ax.yaxis.set_major_locator(MultipleLocator(0.1))

ax.tick_params(axis="both", labelsize=12)
for label in ax.get_xticklabels() + ax.get_yticklabels():
    label.set_fontname("Times New Roman")

legend = ax.legend(fontsize=11, frameon=True)
for text in legend.get_texts():
    text.set_fontname("Times New Roman")

ax.grid(True, linestyle="--", alpha=0.4)

plt.tight_layout()

plt.show()
```
